```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
# Step 2:
points=np.array([
    [3,7],
    [4,6],
    [5,5],
    [6,4],
    [7,3],
    [6,2],
    [7,2],
    [8,4],
    [3,3],
    [2,6],
    [3,5],
    [2,4]
])
```

```python
dbscan=DBSCAN(eps=1.9,min_samples=4)
labels=dbscan.fit_predict(points)
```

```python
plt.figure(figsize=(8,6))
unique_labels=set(labels)     # removes duplicates
for label in unique_labels:
    cluster_points=points[labels==label]
    if cluster_points.size==0:
        continue  # skip empty clusters
    if label==-1:    # noise points
            plt.scatter(cluster_points[:,0],cluster_points[:,1],c='k',label='Noise',marker='x')
    else:
            plt.scatter(cluster_points[:,0],cluster_points[:,1],label=f'Cluster {label}')

plt.legend()
plt.title('DBSCAN Clustering Result')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
print("Cluster labels for each point:")
for i, label in enumerate(labels, start=1):
  print(f"p{i}: Cluster {label if label != -1 else'Noise'}")
```

```python
n_clusters=len(set(labels))-(1 if -1 in labels else 0 )
print("\nNumber of clusters formed (excluding noise ):",n_clusters)
```
