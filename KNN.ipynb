```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
sports=pd.read_csv('/content/drive/MyDrive/Sportsman.csv')
```

```python
sports
```

```python
x=sports[['Height','Weight']]
```

```python
y=sports[['Type']]
```

```python
test_point=[[167,58]]
```

```python
sports=KNeighborsClassifier(n_neighbors=3)
sports.fit(x,y)
```

```python
prediction=sports.predict(test_point)
print("Prediction: ",prediction[0])
```

```python
from re import X
athletes=[x.iloc[i] for i in range(len(x)) if y.iloc[i,0]== 'Athlete']
non_athletes=[x.iloc[i] for i in range(len(x)) if y.iloc[i,0]== 'Non-Athlete']

ath_x=[i[0] for i in athletes]
ath_y=[i[1] for i in athletes]
non_x=[i[0] for i in non_athletes]
non_y=[i[1] for i in non_athletes ]


plt.scatter(ath_x,ath_y,color='blue',label='Athelete')
plt.scatter(non_x,non_y,color='red',label='Non-Athelete')
plt.scatter(test_point[0][0],test_point[0][1],color='green',label='Test Point',marker='X',s=100)

plt.xlabel('Height(cm)')
plt.ylabel('Weight (kg)')
plt.title('K-NN Classification')
plt.legend()
plt.grid(True)
plt.show()
```
