```python
sns.scatterplot(x=titanic['age'],y=titanic['survived'])
```

```python
sns.scatterplot(x='age',y='survived',data=titanic,alpha=0.6,hue='sex')
```

```python
sns.stripplot(x='age',y='survived',data=titanic,jitter=True,hue='sex')
```

```python
sns.swarmplot(x='survived',y='age',data=titanic,hue='sex',dodge=True)
```

```python
sns.swarmplot(x='survived',y='age',data=titanic,hue='sex',dodge=False)
```

```python
plt.hist(titanic['age'],bins=40)
plt.title("HISTOGRAM")
plt.legend()
plt.show()
```

```python
plt.hist(titanic['age'].dropna(),bins=50,color='skyblue',edgecolor='black',label='Age')
plt.title("HISTOGRAM OF PASSANGER AGES")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(titanic['sex'])
```

```python
titanic['survived'].value_counts().plot.pie(autopct='%1.1f%%',colors=['lightcoral','lightgreen'])
plt.title("Survival distribution")
plt.ylabel('')
plt.show()
```

```python
titanic['sex'].value_counts().plot.pie(autopct='%1.1f%%',colors=['red','yellow'])
plt.title("Gender distribution")
plt.ylabel('')
plt.show()
```

```python
sns.boxplot(x=titanic['age'],color='pink')
plt.xlabel("age")
plt.title("Box Plot of Passenger Ages")
plt.show()
```

```python
sns.boxplot(x='survived',y='age',data=titanic,palette='Set2')
plt.title("Age Distribution by Survival")
plt.xlabel("Survived(0=No,1=Yes)")
plt.ylabel("Age")
plt.show()
```
